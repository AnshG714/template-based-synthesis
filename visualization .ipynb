```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
FILES = [
    "rev_list_performance.txt",
    "double_performance_measurement.txt",
    "add1_performance_measurement.txt"
]

performance_map = {}

for file in FILES:
    with open(file, 'r') as f:
        data = f.read()
        performances = data.split("\n")
        l = []
        for performance in performances:
            file_space = performance.find(" ")
            data = [int(el) for el in eval(performance[file_space + 1:])]
            l.append(data)
        performance_map[file] = l            
```

```python
data = np.array(performance_map["rev_list_performance.txt"]).T
fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xticklabels(['minimal', 'less_minimal',
                    'less_less_minimal', 'less_less_less_minimal', 'less_less_less_less_minimal'])
bp = ax.boxplot(data)
plt.show()
```

```python
plt.clf()
data = np.log(performance_map["rev_list_performance.txt"]).T
print(data)
fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xticklabels(['minimal', 'less_minimal',
                    'less_less_minimal', 'less_less_less_minimal', 'less_less_less_less_minimal'])
bp = ax.boxplot(data)
# show plot
plt.show() 
```

```python
def save_plots():
    for program in performance_map:
        p = program[:-4]
        plt.clf()
        data = np.array(performance_map[program]).T
        fig = plt.figure(figsize =(10, 7))
        ax = fig.add_axes([0, 0, 1, 1])
        ax.set_xticklabels(['minimal', 'less_minimal',
                            'less_less_minimal', 'less_less_less_minimal', 'less_less_less_less_minimal'])
        bp = ax.boxplot(data)
        plt.xlabel("Template type")
        plt.ylabel("Performance (milliseconds)")
        plt.title(program)
        plt.savefig(f"{program}.png")
        
        p = program[:-4]
        plt.clf()
        data = np.log(performance_map[program]).T
        fig = plt.figure(figsize =(10, 7))
        ax = fig.add_axes([0, 0, 1, 1])
        ax.set_xticklabels(['minimal', 'less_minimal',
                            'less_less_minimal', 'less_less_less_minimal', 'less_less_less_less_minimal'])
        bp = ax.boxplot(data)
        plt.xlabel("Template type")
        plt.ylabel("Performance (log milliseconds)")
        plt.title(program + "(log)")
        plt.savefig(f"{program}_logarithmic.png")
        
        
save_plots()
```
